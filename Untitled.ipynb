```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "figure.facecolor":"black",
    "axes.facecolor":"black",
    "axes.edgecolor":"orange",
    "xtick.color":"red",
    "ytick.color":"yellow",
    "legend.labelcolor":"linecolor"
})


# Define the analog signal function
def x(t):
    return 2 * np.sin(40 * np.pi * t) - 3 * np.sin(100 * np.pi * t) + np.cos(50 * np.pi * t)

# Define the sampling frequency and duration
fs = 200  # Hz
T = 1  # seconds

# Generate time values for analog signal
t_analog = np.linspace(0, T, 1000)

# Generate time values for sampled signal
t_sampled = np.arange(0, T, 1 / fs)

# Sample the analog signal
x_sampled = x(t_sampled)

# Plot the analog signal
plt.figure(figsize=(10, 5))
plt.plot(t_analog, x(t_analog), label='Analog Signal')
plt.stem(t_sampled, x_sampled, markerfmt='ro', basefmt=" ", linefmt='r', label='Sampled Signal')
plt.title('Analog Signal and Sampled Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
